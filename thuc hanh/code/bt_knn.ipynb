```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
s=np.array([100,120,130,155,180,200,220,240,250,260])
n_bedroom=np.array([3,3,3,4,4,4,5,5,5,6])
price=np.array([250,200,320,370,400,450,500,550,600,620])
```

```python
s = s.reshape(-1, 1)
n_bedroom = n_bedroom.reshape(-1, 1)

# noi du lieu
X = np.concatenate([s, n_bedroom], axis=1)
```

```python
# Chia tập dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, price, test_size=0.2, random_state=42)
```

```python
# Khởi tạo mô hình KNN
knn_regressor = KNeighborsRegressor(n_neighbors=3)
```

```python
# Huấn luyện mô hình
knn_regressor.fit(X_train, y_train)
```

```python
# Dự đoán giá trị mới
new_data_point_1 = np.array([[210, 4]])  
new_data_point_2 = np.array([[300, 2]]) 
new_data_point_3 = np.array([[150, 3]]) 

predicted_price_1 = knn_regressor.predict(new_data_point_1)
predicted_price_2 = knn_regressor.predict(new_data_point_2)
predicted_price_3 = knn_regressor.predict(new_data_point_3)
```

```python
print(f'Gia tri can nha co dien tich 210 m2 + 4 phong ngu: {predicted_price_1[0]:.2f}k($)')
print(f'Gia tri can nha co dien tich 300 m2 + 2 phong ngu: {predicted_price_2[0]:.2f}k($)')
print(f'Gia tri can nha co dien tich 150 m2 + 3 phong ngu: {predicted_price_3[0]:.2f}k($)')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
# Dữ liệu đã có
plt.scatter(n_bedroom, s, label='Dữ liệu đã có', color='blue')

# Dữ liệu cần dự đoán
plt.scatter(new_data_point_1[0, 1], new_data_point_1[0, 0], label='Dữ liệu cần dự đoán', color='red', marker='*')
plt.scatter(new_data_point_2[0, 1], new_data_point_2[0, 0],  color='red', marker='*')
plt.scatter(new_data_point_3[0, 1], new_data_point_3[0, 0],  color='red', marker='*')


# Hiển thị giá nhà trên mỗi điểm dữ liệu
for i in range(len(price)):
    plt.text(n_bedroom[i], s[i], f'{price[i]}k($)', ha='center', va='bottom')

# Hiển thị giá nhà trên điểm dữ liệu cần dự đoán
plt.text(new_data_point_1[0, 1], new_data_point_1[0, 0], f'{predicted_price_1[0]:.2f}k($)', ha='center', va='bottom', color='red')
plt.text(new_data_point_2[0, 1], new_data_point_2[0, 0], f'{predicted_price_2[0]:.2f}k($)', ha='center', va='bottom', color='red')
plt.text(new_data_point_3[0, 1], new_data_point_3[0, 0], f'{predicted_price_3[0]:.2f}k($)', ha='center', va='bottom', color='red')


# Đặt tên trục và tiêu đề biểu đồ
plt.xlabel('Số phòng ngủ')
plt.ylabel('Diện tích')
plt.title('Dữ liệu và Dữ liệu cần dự đoán')

# Hiển thị chú thích
plt.legend()

# Hiển thị biểu đồ
plt.show()
```
