```python
# xoa cac hang ma gia tri cot smoking_status "unknow"
df.drop(df.index[(df["smoking_status"] == "Unknown")],axis=0,inplace=True)
```

```python
# chuan hoa du lieu cac cot du lieu khong phai dang so
from sklearn.preprocessing import LabelEncoder
```

```python
# chuan hoa gioi tinh: male:1
df['gender']=LabelEncoder().fit_transform(df['gender'])

# chuan hoa hon nhan; yes:1
df['ever_married']=LabelEncoder().fit_transform(df['ever_married'])

# chuan hoa cot loai hinhf lam viec: Govt_job:0; Private:1;Self-employed:2
df['work_type']=LabelEncoder().fit_transform(df['work_type'])

#  Urban:1 ; Rural	:0
df['Residence_type']=LabelEncoder().fit_transform(df['Residence_type'])

# formerly smoked:0; never smoked:1; smokes:2 

df['smoking_status']=LabelEncoder().fit_transform(df['smoking_status'])

```
