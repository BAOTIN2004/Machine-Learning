```python
import numpy as np,pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets,linear_model
import seaborn as sns
```

```python
df=pd.read_csv("D:/Machine Learning/thuc hanh/datasets/Salary_Data.csv")
df
```

```python
data=np.array(df)
data
```

```python
years_experience=data[:,0]
salary=data[:,1]
```

```python
plt.figure(figsize=(10, 8))
plt.xlabel('Year_ex')
plt.ylabel('Salary')
plt.scatter(years_experience,salary,color='blue')
plt.show()
```

```python
regr=linear_model.LinearRegression()
```

```python
regr.fit(years_experience.reshape(-1,1),salary)
```

```python
plt.figure(figsize=(10, 8))
plt.xlabel('Year_ex')
plt.ylabel('Salary')
plt.scatter(years_experience,salary,color='black')
plt.plot(years_experience,regr.predict(years_experience.reshape(-1,1)),color='blue')
```

```python
# du doan
year_prediction=5.5
need_prediction=regr.predict(np.array([[year_prediction]]))
plt.figure(figsize=(10, 8))
plt.xlabel('Year_ex')
plt.ylabel('Salary($)')
plt.scatter(years_experience,salary,color='black')
plt.plot(years_experience,regr.predict(years_experience.reshape(-1,1)),color='blue')
plt.scatter([[year_prediction]],[need_prediction],color="red")
plt.annotate(f'Pre {year_prediction}year:{need_prediction[0]: .0f}$',xy=[6.5,need_prediction],xytext=[8.5,need_prediction],arrowprops=dict(facecolor='red',shrink=0.05))
```
